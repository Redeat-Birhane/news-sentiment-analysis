```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import talib
df = pd.read_csv("../data/raw/AAPL.csv")

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

# Ensure numeric types
cols = ["Open", "High", "Low", "Close", "Volume"]
df[cols] = df[cols].apply(pd.to_numeric, errors="coerce")

# Handle missing values
df = df.dropna()

df.head()
```

## 📈 Simple Moving Average (SMA)

```python
df["SMA_20"] = talib.SMA(df["Close"], timeperiod=20)
df["SMA_50"] = talib.SMA(df["Close"], timeperiod=50)
```

## 📈 Exponential Moving Average (EMA)

```python
df["EMA_20"] = talib.EMA(df["Close"], timeperiod=20)
```

## 📊 Relative Strength Index (RSI)

```python
df["RSI"] = talib.RSI(df["Close"], timeperiod=14)
```

## 📉 MACD (Momentum indicator)

```python
macd, signal, hist = talib.MACD(df["Close"],
                               fastperiod=12,
                               slowperiod=26,
                               signalperiod=9)

df["MACD"] = macd
df["MACD_Signal"] = signal
df["MACD_Hist"] = hist
```

## Financial Metrics
Daily Returns

```python
df["Returns"] = df["Close"].pct_change()
```

## Volatility

```python
volatility = df["Returns"].std() * np.sqrt(252)
print("Annualized Volatility:", volatility)
```

## Cumulative Returns

```python
df["Cumulative_Return"] = (1 + df["Returns"]).cumprod()
```

## 📈Visualization

## Price + Moving Averages

```python
plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["Close"], label="Close Price")
plt.plot(df["Date"], df["SMA_20"], label="SMA 20")
plt.plot(df["Date"], df["EMA_20"], label="EMA 20")

plt.title("AAPL Price with Moving Averages")
plt.legend()
plt.show()
```

## RSI Plot

```python
plt.figure(figsize=(12,4))
plt.plot(df["Date"], df["RSI"])
plt.axhline(70, color="red")
plt.axhline(30, color="green")

plt.title("RSI Indicator (Overbought/Oversold)")
plt.show()
```

## MACD Plot

```python
plt.figure(figsize=(12,4))
plt.plot(df["Date"], df["MACD"], label="MACD")
plt.plot(df["Date"], df["MACD_Signal"], label="Signal Line")

plt.legend()
plt.title("MACD Indicator")
plt.show()
```
